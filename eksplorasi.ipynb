```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
from lightgbm import LGBMClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from xgboost import XGBClassifier
%matplotlib inline
```

```python
# Membaca semua file, urutkan dari timestamp terkecil
dfs = []
for file in os.listdir('../dataset/dataset/'):
    dfs.append(pd.read_excel(f'../dataset/dataset/{file}'))
dfs.sort(key=lambda x: x.loc[0, 'timestamp'])
```

```python
for df in dfs:
    print(df)
```

```python
# plot x, y, z dari masing masing sampel
for df in dfs:
    plt.plot(df['timestamp'], df['x'], label='X')
    plt.plot(df['timestamp'], df['y'], label='Y')
    plt.plot(df['timestamp'], df['z'], label='X')
    plt.xlabel('Timestamp')
    plt.ylabel('Accelerometer Value')
    plt.legend()
    plt.show()
```

```python
# Standard scaling x, y, z
temp = dfs[0].copy()
temp[['x','y','z']] = StandardScaler().fit_transform(temp[['x','y','z']])

plt.plot(temp['timestamp'], temp['x'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer X Value')
plt.show()

plt.plot(temp['timestamp'], temp['y'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer Y Value')
plt.show()

plt.plot(temp['timestamp'], temp['z'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer Z Value')
plt.show()
```

```python
# Nilai fluktuasi x, y, z dengan menggunakan asbolute value
temp = abs(temp)

plt.plot(temp['timestamp'], temp['x'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer X Value')
plt.show()

plt.plot(temp['timestamp'], temp['y'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer Y Value')
plt.show()

plt.plot(temp['timestamp'], temp['z'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled Accelerometer Z Value')
plt.show()
```

```python
# correlation antara x, y, z, dan label
master = pd.DataFrame()
for df in dfs:
    master = pd.concat([master, df], ignore_index=True)
master[['x','y','z']] = abs(StandardScaler().fit_transform(master[['x','y','z']].values))
sns.heatmap(master[['x','y','z','label']].corr(), annot=True, vmin=0)
plt.show()
```

```python
# data augmentation di awal dataset
dummy = pd.DataFrame()
for df in dfs:
    dummy = pd.concat([dummy, df], ignore_index=True)
dummy[['x','y','z']] = StandardScaler().fit_transform(dummy[['x','y','z']].values)

mean_x = np.mean(dummy[dummy['label']==0]['x'].values)
mean_y = np.mean(dummy[dummy['label']==0]['y'].values)
mean_z = np.mean(dummy[dummy['label']==0]['z'].values)

std_x = np.std(dummy[dummy['label']==0]['x'].values)
std_y = np.std(dummy[dummy['label']==0]['y'].values)
std_z = np.std(dummy[dummy['label']==0]['z'].values)

timestamps = [i for i in range(np.int64(dummy.iloc[0]['timestamp'])-60000, np.int64(dummy.iloc[0]['timestamp']), 40)]
noise_x = np.random.normal(mean_x, std_x, len(timestamps))
noise_y = np.random.normal(mean_y, std_y, len(timestamps))
noise_z = np.random.normal(mean_z, std_z, len(timestamps))

dummy = pd.concat([pd.DataFrame({'timestamp':timestamps, 'x':noise_x, 'y':noise_y, 'z':noise_z, 'label':[0 for _ in range(len(timestamps))]}), dummy], ignore_index=True)
plt.plot(dummy['timestamp'], dummy['x'])
plt.xlabel('Timestamp')
plt.ylabel('Standard scaled x value')
plt.show()
plt.plot(dummy['timestamp'], dummy['y'])
plt.xlabel('Timestamp')
plt.ylabel('Standard scaled y value')
plt.show()
plt.plot(dummy['timestamp'], dummy['z'])
plt.xlabel('Timestamp')
plt.ylabel('Standard scaled z value')
plt.show()
```

```python
# convert x, y, z menjadi nilai absolute-nya
master = dummy.copy()
master = abs(master)

plt.plot(master['timestamp'], master['x'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled x value')
plt.show()
plt.plot(master['timestamp'], master['y'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled y value')
plt.show()
plt.plot(master['timestamp'], master['z'])
plt.xlabel('Timestamp')
plt.ylabel('Scaled z value')
plt.show()
```

```python
# create r_mean feature
for i in range(0, 30, 10):
    for j in range(10, 50, 10):
        master[f'r_mean_x_{i}_{j}'] = master['x'].shift(i * 1000 // 40).rolling(j * 1000 // 40).mean()
        master[f'r_mean_y_{i}_{j}'] = master['y'].shift(i * 1000 // 40).rolling(j * 1000 // 40).mean()
        master[f'r_mean_z_{i}_{j}'] = master['z'].shift(i * 1000 // 40).rolling(j * 1000 // 40).mean()

# plt.plot(master['timestamp'], master['r_mean_x_0_20'], label='Rolling Mean 20s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
# plt.plot(master['timestamp'], master['r_mean_x_20_20'], label='Rolling Mean 40s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
# plt.plot(master['timestamp'], master['r_mean_x_40_20'], label='Rolling Mean 60s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
```

```python
# create r_std feature
for i in range(0, 30, 10):
    for j in range(10, 50, 10):
        master[f'r_std_x_{i}_{j}'] = master['x'].shift(i * 1000 // 40).rolling(j * 1000 // 40).std()
        master[f'r_std_y_{i}_{j}'] = master['y'].shift(i * 1000 // 40).rolling(j * 1000 // 40).std()
        master[f'r_std_z_{i}_{j}'] = master['z'].shift(i * 1000 // 40).rolling(j * 1000 // 40).std()
        
# plt.plot(master['timestamp'], master['r_std_x_0_20'], label='Rolling Std 20s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
# plt.plot(master['timestamp'], master['r_std_x_20_20'], label='Rolling Std 40s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
# plt.plot(master['timestamp'], master['r_std_x_40_20'], label='Rolling Std 60s')
# plt.plot(master['timestamp'], master['label'], label='Label')
# plt.xlabel('Timestamp')
# plt.legend()
# plt.show()
```

```python
# save master data
master.to_csv('../dataset/dummy/DUMMY.csv', index=False)

# split data
train_test = master.iloc[1500:]
val = train_test[train_test['timestamp'].isin(dfs[-1]['timestamp'])]
train = train_test.iloc[:-val.shape[0]]

train_y = train['label']
# train_x = train.drop(columns=['label', 'r_std_x_20', 'r_std_x_40', 'r_std_x_60', 'r_std_y_20', 'r_std_y_40', 'r_std_y_60', 'r_std_z_20', 'r_std_z_40', 'r_std_z_60'])
train_x = train.drop(columns='label')

val_y = val['label']
# val_x = val.drop(columns=['label', 'r_std_x_20', 'r_std_x_40', 'r_std_x_60', 'r_std_y_20', 'r_std_y_40', 'r_std_y_60', 'r_std_z_20', 'r_std_z_40', 'r_std_z_60'])
val_x = val.drop(columns='label')
```

```python
clf = LGBMClassifier()
clf.fit(train_x, train_y)
pred = clf.predict(val_x)

accuracy_score(val_y, pred)
```

```python
plt.plot(val_x['timestamp'], pred)
plt.plot(val_x['timestamp'], val_y)
plt.show()
```

```python
clf = HistGradientBoostingClassifier()
clf.fit(train_x, train_y)
pred = clf.predict(val_x)

accuracy_score(val_y, pred)
```

```python
plt.plot(val_x['timestamp'], pred)
plt.plot(val_x['timestamp'], val_y)
plt.show()
```
